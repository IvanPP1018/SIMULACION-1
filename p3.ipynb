```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt

def u(n):
  lista=[]

  for i in range (n):
    u=rnd.random()
    if u < 0.15:
      x=1
    elif u<0.20:
      x=2
    elif u<0.55:
      x=3
    else:
      x=4
    lista.append(x)
  return lista
```

```python

G=u(100)
media=np.mean(G)
varianza= np.var(G)
plt.hist(G)
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y=cdf(G)
plt.step(X,Y)
```
