```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt

def expo(landa,n):
  L= -1/landa
  lista=[]
  for i in range(n):
    u=rnd.random()
    x=L*np.log(u)
    lista.append(x)
  return lista

```

```python
expo(2,10)
```

```python
G=expo(0.5,18)



```

```python
media=np.mean(G)
print(media)
```

```python
varianza=np.var(G)
print(varianza)
```

```python
plt.hist(G)
```

```python
plt.plot(G)
```

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y=cdf(G)
```

```python
plt.step(X,Y)
```

```python
media=np.mean(G)
plt.plot(X,Y)
plt.axvline(media,color='r',linestyle='----')
plt.axhline(0.5,color='r',linestyle='----')
```
